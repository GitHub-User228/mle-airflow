```python
import psycopg2
from psycopg2 import sql

# Параметры подключения к PostgreSQL
host = "rc1b-uh7kdmcx67eomesf.mdb.yandexcloud.net"
port = 6432
user = "playground_mle_20240730_73c4e0c760"
password = "HI&ykgu6tj"
dbname = "playground_common"  # Подключаемся к существующей базе данных, например, postgres

# Название новой базы данных
new_dbname = "playground_common2"

try:
    # Подключаемся к PostgreSQL серверу
    conn = psycopg2.connect(
        host=host,
        port=port,
        user=user,
        password=password,
        dbname=dbname
    )
    conn.autocommit = True

    # Создаем курсор для выполнения SQL-запросов
    cursor = conn.cursor()

    # Создаем новую базу данных
    cursor.execute(sql.SQL("CREATE DATABASE {}").format(
        sql.Identifier(new_dbname))
    )

    print(f"Database '{new_dbname}' created successfully.")

except psycopg2.Error as e:
    print(f"Error creating database: {e}")

finally:
    # Закрываем курсор и соединение
    if cursor:
        cursor.close()
    if conn:
        conn.close()
```
